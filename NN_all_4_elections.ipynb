```python
import pandas as pd
import json
import os
```

```python
ru_lists_all = pd.read_csv("data/RU_lists.csv", encoding='utf8', sep=';')
ru_lists_all.apply(lambda x: pd.api.types.infer_dtype(x.values))
ru_lists_nn = ru_lists_all.loc[ru_lists_all["reg"] == "Нижегородская область"]
```

```python
with open("data/NN_regional_lists.json", encoding = 'utf8') as file:
    nn_lists_json = json.load(file)
nn_lists_list = []
for line in nn_lists_json:
    result_lines = {}
    for protocol_line in line['protocol_lines']:
        result_lines[protocol_line['line_name']] = int(protocol_line['line_value'])
    export_line = dict(result_lines)
    export_line.update({'uik_name': line['UIK'], 'uik_number': int(line['UIK_namber']), 'url':line['url']})
    nn_lists_list.append(export_line)
nn_lists = pd.DataFrame(nn_lists_list)
nn_lists.head()
```

```python
with open("data/NN_regional_persons.json", encoding = 'utf8') as file:
    nn_lists_json = json.load(file)
nn_persons_list = []
for line in nn_lists_json:
    result_lines = {}
    for protocol_line in line['protocol_lines']:
        result_lines[protocol_line['line_name']] = int(protocol_line['line_value'])
    export_line = dict(result_lines)
    export_line.update({'uik_name': line['UIK'], 'uik_number': int(line['UIK_namber']), 'url':line['url']})
    nn_lists_list.append(export_line)
nn_persons = pd.DataFrame(nn_lists_list)
nn_persons.head()
```
